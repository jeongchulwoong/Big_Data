```python
# Naver 에서 특정 키워드로 기사 검색하기

from selenium import webdriver
from selenium.webdriver.common.by import By
import time


query_txt = input('키워드를 입력하세요: ')
#query_txt = '뉴스', '축구', '빅데이터' 을 넣었을 때만 해보았습니다. 키워드를 입력시 뉴스 데이터만 출력
print("\n")

#Step 3. 크롬 드라이버 설정 및 웹 페이지 열기
chrome_driver = 'C:\py_temp/chromedriver'
driver = webdriver.Chrome()

url = 'https://www.naver.com/'
driver.get(url)                  # 접속 명령어
time.sleep(5)                    # 5초 설정(인터넷 환경에 따라)

#Step 4. 자동으로 검색어 입력 후 조회하기
element = driver.find_element(By.ID, 'query')
driver.find_element(By.ID, 'query').click( )
element.send_keys(query_txt)                # 글자를 타이핑하라
element.send_keys("\n")                     # enter키를 쳐라
```

```python
from bs4 import BeautifulSoup

# BeautifulSoup로 본문 내용만 추출하기
html_1 = driver.page_source
soup = BeautifulSoup(html_1, 'html.parser')

headlines = []
contents = []

news_boxes = soup.find_all('div', class_='news_contents')

for news_box in news_boxes :
    headline = news_box.find('a', class_='news_tit')['title']
    content = news_box.find('div', class_='news_dsc').text.strip()

    headlines.append(headline)
    contents.append(content)

for i in range(len(headlines)):
    print(f"Headline {i+1}: {headlines[i]}\n")
    print(f"Content {i+1}: {contents[i]}\n\n")


```
