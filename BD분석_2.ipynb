```python
# Okt 형태소 분석기 생성
from konlpy.tag import Okt
okt = Okt()
```

```python
sentence1 = okt.pos(text)               #pos로 토큰화
print(sentence1)
```

```python
from collections import Counter           # 빈도수를 카운트 라이브러리
import collections

adj_list = []
for word, tag in sentence1:                # sentence1의 단어 태그 word, tag 저장
    if tag in ['Noun', 'Adjective']:      # tag의 명사와 형용사만 추출
        adj_list.append(word)             # adj_list에 추출된 word를 추가

counts = collections.Counter(adj_list)    # adj_list를 counts에 저장
tag1 = counts.most_common()                # 가장 많은 빈도수를 tag에 저장
print(tag1)
```

```python
from wordcloud import WordCloud

wc = WordCloud(font_path = font_path,
               background_color='white',
               max_font_size=100,
               width=500, height=500)

cloud = wc.generate_from_frequencies(dict(tag1))

import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
plt.imshow(cloud)
```
