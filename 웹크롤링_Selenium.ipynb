```python
# riss.kr 에서 특정 키워드로 논문 / 학술 자료 검색하기

#Step 1. 필요한 모듈을 로딩합니다
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
```

```python
#Step 2. 사용자에게 검색 관련 정보들을 입력 받습니다.
print("=" *100)
print(" 이 크롤러는 RISS 사이트의 논문 및 학술자료 수집용 웹크롤러.")
print("=" *100)
query_txt = input('수집할 자료의 키워드?(여러개일 경우 , 로 구분하여 입력): ')
#query_txt = '해양자원,도시재생'
print("\n")
```

```python
#Step 3. 크롬 드라이버 설정 및 웹 페이지 열기
chrome_path = "C:/py_temp/chromedriver.exe"
#chrome_driver = '/Users/yeomd/anaconda3/chromedriver'
#driver = webdriver.Chrome(chrome_path)
driver = webdriver.Chrome()


url = 'http://www.riss.kr/'
driver.get(url)                  # 접속 명령어
time.sleep(5)                    # 5초 설정(인터넷 환경에 따라)
```

```python
#Step 4. 자동으로 검색어 입력 후 조회하기
#element = driver.find_element_by_id("query)  #by_name, by_xpath(....)
element = driver.find_element(By.ID, 'query')
#driver.find_element_by_id("query").click( )
driver.find_element(By.ID, 'query').click( )
element.send_keys(query_txt)                # 글자를 타이핑하라
element.send_keys("\n")                     # enter키를 쳐라
```
